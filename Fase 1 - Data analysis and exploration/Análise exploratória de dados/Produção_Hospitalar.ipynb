```python
# DESAFIOS: reposicionar a legenda
#           melhorar a visualização
#           adicionar título
#           adicionar labels nos eixos

uf = dados_usaveis.T.columns.to_list()

fig, ax = plt.subplots(figsize=(20,10), layout='constrained')

for i in range(0, len(dados_usaveis)):
   ax.plot(dados_usaveis.T.index, dados_usaveis.T.iloc[:, i], label=uf[i])
   
ax.set_title("Gastos com Saúde por Unidade da Federação ao Longo do Tempo")
ax.set_xlabel("Meses")
ax.set_ylabel("Gastos com Saúde")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.set_xticks(['2007/Ago', '2009/Set', '2011/Out', '2013/Nov', '2015/Dez', '2018/Jan', '2020/Fev', '2022/Mar'])

pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width, pos.height * 0.85])


ax.legend(
    loc='upper center', 
    bbox_to_anchor=(0.5,0),
    ncol=9,
    borderaxespad=5
)


plt.show()
```

```python
dados_usaveis["Total"] = dados_usaveis.sum(axis = 1)
dados_usaveis
```

```python
# DESAFIO: adicionar uma coluna com a região de cada estado
#          ordenar o dataframe pelo total          
#          adicionar o seu estado na lista de 7 estados

dados_usaveis.insert(0, "Região", "")
dados_usaveis
```

```python
dados_usaveis.iloc[:7 , 0] = "Norte"
dados_usaveis.iloc[7 : 16, 0] = "Nordeste"
dados_usaveis.iloc[16: 20, 0] = "Sudeste"
dados_usaveis.iloc[20 : 23, 0] = "Sul"
dados_usaveis.iloc[23:, 0] = "Centro Oeste"
```

```python
dados_usaveis.head()
```

```python
dados_usaveis.sort_values(by='Total', ascending=False, inplace=True)
dados_usaveis.head()
```

```python
dados_sp = dados_usaveis.iloc[0, :]

df_dados_sp = pd.DataFrame(dados_sp.values, index=dados_sp.index, columns= ['35 São Paulo']).T
df_dados_sp.index.name= 'Unidade da Federação'

df_dados_sp
```
