````python
from datetime import date

meses = {
    "Jan" : 1,
    "Fev" : 2,
    "Mar" : 3,
    "Abr" : 4,
    "Mai" : 5,
    "Jun" : 6,
    "Jul" : 7,
    "Ago" : 8,
    "Set" : 9,
    "Out" : 10,
    "Nov" : 11,
    "Dez" : 12
}

def para_dia(ano_mes: str):
    ano: int = int(ano_mes[:4])
    nome_mes: str = ano_mes[5:]
    mes : int = meses[nome_mes]
    return date(ano, mes, 1)
````
