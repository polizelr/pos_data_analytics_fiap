```python
def insere_gastos_e_gastos_por_habitante(ordenados_por_total, gastos_e_populacao, mes):
    gastos = ordenados_por_total[mes]    
    gastos_e_populacao[f"{mes}_gastos"] = gastos * 1_000_000
    gastos_e_populacao[f"{mes}_gasto_por_habitante"] = gastos_e_populacao[f"{mes}_gastos"] / gastos_e_populacao["populacao"]
```

```python
def soma_meses_do_ano (ordenados_por_total, gastos_e_populacao, ano):
    meses = ordenados_por_total.columns[ordenados_por_total.columns.str.contains(f"{ano}")]
    gastos_e_populacao[f"{ano}_gastos"] = ordenados_por_total[ordenados_por_total.columns[ordenados_por_total.columns.str.contains(f"{ano}")]].sum(axis=1) * 1_000_000
```
